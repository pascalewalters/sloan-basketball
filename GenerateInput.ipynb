```python
import os
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 500)
pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
# read in raw data

pbp_dfs = []
shots_dfs = []
poss_dfs = []

fp = "ExploreSTData/playbyplay/"
for f in os.listdir(fp):
    if f != ".ipynb_checkpoints" and "M_" in f:
        #pbp_dfs.append(pd.read_csv(fp + f + "/" + f + ".csv"))
        pbp_df = pd.read_csv(fp + f + "/" + f + ".csv")
        pbp_df["GameId"] = f
        pbp_dfs.append(pbp_df)
        
        shots_df = pd.read_csv("ExploreSTData/timeseries/" + f + "/" + f + "_shots.csv")
        shots_df["GameId"] = f
        shots_dfs.append(shots_df)
        
        poss_dfs.append(pd.read_csv("ExploreSTData/timeseries/" + f + "/" + f + "_playerPossession.csv"))

pbp = pd.concat(pbp_dfs)
shots = pd.concat(shots_dfs)
poss = pd.concat(poss_dfs)
```

```python
misses = shots.loc[(shots.sessionState=="LIVE") & (shots.event == "enrichedmiss")].reset_index(drop=True)
misses
```

```python
# Pascale's filtered Rebounds data
rebounds = pd.read_csv("rebounds.csv", index_col=0).reset_index(drop=True)
rebounds
```

```python
# basic data validation (get the shot before the rebound)

tkShots = []
tkRebounds = []
for i, row in rebounds.iterrows():
    
    #print(i)
    #print(row.to_dict())
    possibles = misses.loc[(misses.GameId == row["game"]) & (misses.timestamp <= row["Timestamp"])]
    shot = possibles.iloc[-1]
    #print(shot.to_dict())
    #print(possibles.index)
    #print(possibles.index[0])
    diff = row["Timestamp"] - shot["timestamp"]
    #print(diff)
    
    # check that the times actually match up (super basic error checking; would do more with more time)
    #if diff > 5000:
    #    toDrop.append(i)
    #else:
    #    toKeep.append(possibles.index[-1])
    if diff < 5000:
        tkShots.append(possibles.index[-1])
        tkRebounds.append(i)

misses = misses.loc[tkShots]
rebounds = rebounds.loc[tkRebounds]
```
